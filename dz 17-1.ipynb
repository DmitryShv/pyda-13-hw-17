```python
# попробовал перевести данные в int из float
# это не сработало, так как в данных имеются пропуски
df_horse.surgery = df_horse.surgery.astype(int)
```

```python
# посмотрим на уникальные значения в каждом столбце
# в столбце "age" видим "9", но не видим "2", которая должна быть, судя по описанию набора данных. Это можно назвать выбросом
# и требуется заменить, привести к соответствующему виду. 
# остальные данные, на первый взляд, соответствуют описанию и не имеют выбросов
df_horse.value_counts()
```

```python
# заменим 9 на 2
df_horse.age.replace(9, 2, inplace=True)
```

```python
df_horse.age.unique()
```

```python
# подсчет пропусков. Стандартная форма
# посмотрим где пропуски еще раз. И подумаем что с этим делать
# видим, что surgery и outcome имеют по 0.3% пропусков, но они для нас значимы
# так как говорят о типе вмешательстве и итоговом результате. Я принимаю решение такие данные удалить
# ведь, если мы не можем получить информацию из других источников, то на основании этого фрейма достроить не сможем
for col in df_horse.columns:
    pct_missing = df_horse[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
# удалим строки без данных в столбцах 'surgery', 'outcome'
# видим, что количество строк 299 вместо 300, значит пропущенные данные были в одной и той же строке.
df_horse.dropna(subset=['surgery', 'outcome'], inplace=True)
df_horse
```

```python
# теперь можно перевести в другой тип данных, если потребуется
# проверяем
df_horse.surgery = df_horse.surgery.astype(int)
```

```python
# да, все сработало
df_horse
```

```python
df_horse.quantile([0.25, 0.75])
```

```python
q1 = df_horse.rectal_temperature.quantile(0.25)
q3 = df_horse.rectal_temperature.quantile(0.75)
iqr = q3 - q1
print(q1, q3, iqr)
lower_bound = q1 - (1.5 * iqr) # это формула, ее надо просто запомнить :)
print(lower_bound)
upper_bound = q3 + (1.5 * iqr)
print(upper_bound)
remove_outliers = df_horse[df_horse.rectal_temperature.between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal_temperature', ascending=False)
remove_outliers
```

```python
# что произойдет, если мы удалим все строки, в которых есть хотя бы один пропуск?
# удалится почти половина фрейма
df_horse_not_drop = df_horse.dropna() #inplace=True
df_horse_not_drop
```
